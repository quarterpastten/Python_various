## Westside Hospital (Breast Cancer)

```python
import pandas as pd 
import numpy as np 
import sklearn
import statsmodels.api as sm  
```

```python
df = pd.read_csv('LSE_DA301_Week_2_files/Data/breast_cancer_data.csv')
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
# drop null val columns and the id column:
df = df.drop(['Unnamed: 32', 'id'], axis=1)

# -----------------------------------------------------
# check if the d ata is balanced (used target var)
# -----------------------------------------------------
import seaborn as sns

sns.set_theme(style='darkgrid')
ax = sns.countplot(x='diagnosis', data=df)
ax.set_title('Target Imbalance')

# data is balanced enough 
```

```python
feature_cols=[c for c in df.columns if c!= 'diagnosis']
X = df[feature_cols]
y = df['diagnosis']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=0)
from sklearn.tree import DecisionTreeClassifier
cls = DecisionTreeClassifier(random_state=0)
cls.fit(X_train,y_train)
y_predict = cls.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
confusion_matrix = confusion_matrix(y_test, y_predict, labels=['M', 'B'])  
print(confusion_matrix)
sns.heatmap(confusion_matrix, annot=True, fmt='g')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```
