```python
import pandas as pd
import seaborn as sns
movies = pd.read_csv("LSE_DA201_Week_4_files/LSE_DA201_Week_4_files/Data/movies.csv")
ott = pd.read_excel("LSE_DA201_Week_4_files/LSE_DA201_Week_4_files/Data/ott.xlsx")
ott.shape
```

```python
# merge the 2 dataframes
movies_ott = pd.merge(movies, ott, on='ID', how='inner')
movies_ott.head()
```

```python
# plot a countplot
sns.countplot(x='Age', data=movies_ott, hue='Netflix')
```

```python
# plot a histogram
sns.histplot(data=movies_ott, x='IMDb', binwidth=.5)
```

```python
# plot a scatterlot
sns.scatterplot(x='Rotten Tomatoes', y='IMDb', data=movies_ott)
```

```python
# box plot
sns.boxplot(data=movies_ott, x='Age', y='IMDb')
```
