```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
bc_data = pd.read_csv('LSE_DA301_Week_2_files/Data/breast_cancer_data.csv')

bc_data.head()
```

```python
bc_data.info()
```

```python
bc_data.describe()
```

```python
bc_data = bc_data.drop(['Unnamed: 32', 'id'], axis=1)
```

```python
# -----------------------------------------------------
# check if the d ata is balanced (used target var)
# -----------------------------------------------------
import seaborn as sns

sns.set_theme(style='darkgrid')
ax = sns.countplot(x='diagnosis', data=bc_data)
ax.set_title('Target Imbalance')
```

```python
# ---------------------------
# build a BLR model
# ---------------------------
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split 

X = bc_data.drop(['diagnosis'], axis=1)
y = bc_data['diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=0)
logreg = LogisticRegression() 

logreg.fit(X_train, y_train) 
```

```python
X
```

```python
# ----------------------------
# Check accuracy of model
# ----------------------------
y_pred = logreg.predict(X_test)
y_pp = logreg.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix  

confusion_matrix = confusion_matrix(y_test, y_pred, labels=['M', 'B'])  
sns.heatmap(confusion_matrix, annot=True, fmt='g')
```

```python
confusion_matrix
```

```python
confusion = pd.DataFrame(confusion_matrix, index=['predicted_cancer', 'predicted_healthy'], \
                         columns=['is_cancer', 'is_healthy'])
confusion
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```

```python
# Import the necessary package.
from sklearn.metrics import classification_report  

# Print a report on the model's accuracy.
print(classification_report(y_test, y_pred)) 
```
